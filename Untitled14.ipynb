````python
# Scanning all the files:
def detectingAndReadingFile(file_path):
    ext = file_path.split('.')[-1]
    if ext == 'csv':
        return pd.read_csv(file_path)
    elif ext == 'xlsx':
        return pd.read_excel(file_path)
    elif ext == 'txt':
        with open(file_path, 'r') as f:
            return f.read()
    elif ext == 'pdf':
        with pdfplumber.open(file_path) as pdf:
            return "\n".join([page.extract_text() for page in pdf.pages if page.extract_text()])
    elif ext == 'docx':
        doc = Document(file_path)
        return "\n".join([para.text for para in doc.paragraphs])
    elif ext in ['png', 'jpg', 'jpeg']:
        image = Image.open(file_path)
        return pytesseract.image_to_string(image)
    else:
        return "Unsupported file format"
````

````python
# Function: Interact with LLM
def generatingResponseFromLLM(question, context=""):
    prompt = f"""
You are a data analyst agent. Analyze the data/text below and answer the user question.

Data/Content:
{context[:3000]}  # Limit to 3000 characters if needed

Question: {question}
Answer:"""

    response = together.Complete.create(
        prompt=prompt,
        model=model_name,
        max_tokens=500,
        temperature=0.7,
        top_k=1,
        top_p=0.9,
    )
    # if hasattr(response, 'choices') and len(response.choices) > 0:
    #   return response.choices[0].text # Adjusted to access .text directly
    # else:
    #     print("Warning: Could not retrieve text from LLM response.")
    #     print(f"Full response object: {response}")
    #     return "Error: Could not get a valid response from the LLM."
    return response['choices'][0]['text']
````

````python
# prompt: I want a function which gives prompt to the llm for the generation of the graph or plot matplotlib code based on the user input of question

def generatingCodeForTheGraph(question, dataframe_variable_name='df'):
    if not isinstance(parsed, pd.DataFrame):
        return "Error: The uploaded content is not a pandas DataFrame. Cannot generate plot code."

    # Get column names and a few data types
    column_info = parsed.dtypes.to_string()
    column_names = list(parsed.columns)

    prompt = f"""
You are a Python data visualization expert. The user has provided a question about visualizing data from a pandas DataFrame named '{dataframe_variable_name}'. Your task is to generate the complete Python code using matplotlib to create the requested plot.

The DataFrame '{dataframe_variable_name}' has the following columns and data types:
{column_info}

The column names are: {column_names}

Based on the user's question and the available columns, generate the Python code using matplotlib to create the plot.
- Import necessary libraries (like matplotlib.pyplot).
- Assume the DataFrame is already loaded into a variable named '{dataframe_variable_name}'.
- Do not include any markdown or explanations, just the executable Python code.
- Make sure the plot code is complete and can be run directly.
- Add comments to explain key parts of the code.
- If the question is unclear or impossible with the available data, return a brief comment explaining why.

User Question: {question}

Python Plot Code:
"""
    return prompt
````

````python
import re
#Since llm also provide comments inside the generated code for the plot and which causes error there i have written this
#Function to clean it using regex and then use it for plot execution.
def cleanPlotCode(code):
    # Remove markdown code blocks (```python ... ```)
    code = re.sub(r'```.*?\n(.*?)```', r'\1', code, flags=re.DOTALL)
    # Remove any stray system-level comments
    code = re.sub(r'<.*?>', '', code)
    return code.strip()
````

````python

def generateGraphOrAnswer(data_content):
    while True:
        print("\nWhat would you like to do?")
        print("1. Ask a question about the data/content")
        print("2. Generate a plot or graph (only if data is a DataFrame)")
        print("3. Exit")

        choice = input("Enter your choice (1, 2, or 3): ")

        if choice == '1':
            question = input("What is your question? ")
            answer = generatingResponseFromLLM(question, context=data_content)
            print("\nLLM Answer:")
            print(answer)
            # Follow-up questions loop
            while True:
                follow_up = input("Do you have a follow-up question? (yes/no): ").lower()
                if follow_up == 'yes':
                    question = input("What is your follow-up question? ")
                    answer = generatingResponseFromLLM(question, context=data_content + "\nPrevious interaction:" + answer) # Include previous interaction for context
                    print("\nLLM Answer:")
                    print(answer)
                else:
                    break

        elif choice == '2':
            if isinstance(data_content, pd.DataFrame):
                plot_question = input("Describe the plot you want to create (e.g., 'create a scatter plot of column A vs column B'): ")
                plot_prompt = generatingCodeForTheGraph(plot_question, dataframe_variable_name='parsed')
                print("\nGenerating plot code...")
                plot_code = cleanPlotCode(generatingResponseFromLLM(plot_prompt))

                print("\nGenerated Plot Code:")
                print(plot_code)
                execute_code = input("Would you like to execute this code? (yes/no): ").lower()
                if execute_code == 'yes':
                    try:
                        # Executing the generated plot code. 'parsed' is already available.
                        exec(plot_code, {'parsed': data_content, 'plt': plt, 'sns': sns, 'pd': pd})
                        plt.show()
                    except Exception as e:
                        print(f"Error executing plot code: {e}")
                else:
                    print("Plot code execution skipped.")
            else:
                print("Plot generation is only available for DataFrame content.")

        elif choice == '3':
            print("Exiting interactive session.")
            break

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")


````
